```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

# Set plotting style
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 13,
    "axes.linewidth": 1.3,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 5,
    "ytick.major.size": 5,
    "xtick.minor.size": 3,
    "ytick.minor.size": 3,
    "text.usetex": False,
})

# --- Morse parameters (Constants for 16O2 SR system) ---
re_X, De_X, a_X = 1.207, 41256., 2.680
re_B, De_B, a_B = 1.604,  8110., 1.800
B_min  = 46500.
B_diss = B_min + De_B   # ≈ 54610 cm⁻¹

def V_X(R): return De_X * (1 - np.exp(-a_X * (R - re_X)))**2
def V_B(R): return B_min + De_B * (1 - np.exp(-a_B * (R - re_B)))**2

def E_vib_X(v): return 1580.*(v+0.5) - 12.0*(v+0.5)**2
def E_vib_B(v): return B_min + 700.*(v+0.5) - 15.1*(v+0.5)**2

def turning_points_X(Ev):
    s = np.sqrt(Ev / De_X)
    return re_X - np.log(1 + s) / a_X, re_X - np.log(1 - s) / a_X

def turning_points_B(Ev):
    s = np.sqrt((Ev - B_min) / De_B)
    return re_B - np.log(1 + s) / a_B, re_B - np.log(1 - s) / a_B

# --- Figure layout ---
fig = plt.figure(figsize=(12, 7.5))
gs  = gridspec.GridSpec(1, 2, width_ratios=[3.2, 1.4], wspace=0.10)
ax  = fig.add_subplot(gs[0])
axr = fig.add_subplot(gs[1], sharey=ax)

R  = np.linspace(0.82, 4.5, 8000)
Vx = V_X(R)
Vb = V_B(R)

E_lo, E_hi = -2500, 61000

# --- Shading: only around / above dissociation limit ---
ax.axhspan(B_diss - 3500, B_diss, color='#d6eaf8', alpha=0.60, zorder=0)
ax.axhspan(B_diss, E_hi, color='#d5f5e3', alpha=0.55, zorder=0)

# Dissociation limit line
ax.axhline(B_diss, color='#4a7c59', lw=1.4, ls='--', alpha=0.80, zorder=3)
ax.text(3.90, B_diss + 500, 'Dissociation limit', color='#4a7c59', fontsize=9, va='bottom', ha='right', style='italic')

# Continuum annotation
ax.text(3.40, B_diss + 3200, 'Continuum\n(unbound)', fontsize=9, color='#4a7c59', va='center', ha='center', style='italic',
        bbox=dict(fc='#eafbf1', ec='#4a7c59', lw=0.7, boxstyle='round,pad=0.35', alpha=0.93))

# --- Level sticks: X state, v''=0..5 ---
for v in range(6):
    Ev   = E_vib_X(v)
    Ri, Ro = turning_points_X(Ev)
    lw   = 1.8 if v == 0 else 1.1
    al   = max(0.35, 0.85 - v * 0.08)
    ax.plot([Ri, Ro], [Ev, Ev], color='#2b6bbd', lw=lw, ls='-', alpha=al, zorder=4, solid_capstyle='round')
    
    # NEW: Annotation for v''=0
    if v == 0:
        ax.text(Ro + 0.05, Ev, r"$v''=0$", fontsize=10, color='#2b6bbd', va='center', ha='left', fontweight='bold')

# --- Level sticks: B state ---
for v in range(22):
    Ev = E_vib_B(v)
    if Ev >= B_diss:
        break
    Ri, Ro = turning_points_B(Ev)
    lw   = 1.6 if v == 0 else 1.1
    al   = max(0.25, 0.78 - v * 0.04)
    ax.plot([Ri, Ro], [Ev, Ev], color='#8b1a1a', lw=lw, ls='-', alpha=al, zorder=4, solid_capstyle='round')

# --- Vertical excitation arrow ---
v_X_ex = 0
v_B_ex = 4
E_X_ex = E_vib_X(v_X_ex)
E_B_ex = E_vib_B(v_B_ex)
R_arrow = re_X + 0.045

ax.annotate('', xy=(R_arrow, E_B_ex), xytext=(R_arrow, E_X_ex),
            arrowprops=dict(arrowstyle='->', color='#226633', lw=2.3, mutation_scale=18), zorder=7)
ax.text(R_arrow + 0.055, (E_X_ex + E_B_ex) / 2, r'$h\tilde{\nu}$', 
        fontsize=13, color='#226633', va='center', ha='left', style='italic', fontweight='bold')

# --- PEC curves (Increased thickness) ---
ax.plot(R[Vx < 43500], Vx[Vx < 43500], color='#1a3a6b', lw=4.5, zorder=6, solid_capstyle='round')
mask_B = (Vb < E_hi) & (R > 1.07)
ax.plot(R[mask_B], Vb[mask_B], color='#8b1a1a', lw=4.5, zorder=6, solid_capstyle='round')

# --- State labels ---
ax.text(2.80, 6500,  r'X $^3\Sigma_g^-$', color='#1a3a6b', fontsize=13, fontweight='bold', va='center')
ax.text(2.80, B_min + 2600, r'B $^3\Sigma_u^-$', color='#8b1a1a', fontsize=13, fontweight='bold', va='center')

# --- ∆re annotation ---
y_dre = B_min - 4200
ax.annotate('', xy=(re_B, y_dre), xytext=(re_X, y_dre),
            arrowprops=dict(arrowstyle='<->', color='black', lw=1.6, mutation_scale=11))
ax.text((re_X + re_B) / 2, y_dre - 1400, r'$\Delta r_e = 0.397\ \mathrm{\AA}$', 
        ha='center', fontsize=10.5, color='black', style='italic')
ax.plot([re_X, re_X], [E_lo, y_dre], color='#1a3a6b', lw=0.8, ls=':', alpha=0.35, zorder=2)
ax.plot([re_B, re_B], [B_min, y_dre], color='#8b1a1a', lw=0.8, ls=':', alpha=0.35, zorder=2)

ax.set_xlabel(r'Internuclear distance $R\ (\mathrm{\AA})$', fontsize=14)
ax.set_ylabel(r'Energy $(\mathrm{cm}^{-1})$', fontsize=14)
ax.set_xlim(0.82, 4.15)
ax.set_ylim(E_lo, E_hi)
ax.tick_params(which='major', direction='in', length=6, top=True, right=True)
ax.tick_params(which='minor', direction='in', length=3, top=True, right=True)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_major_locator(MultipleLocator(10000))
ax.text(0.015, 0.975, '(a)', transform=ax.transAxes, fontsize=13, fontweight='bold', va='top')

# --- Right panel: cross-section illustration ---
axr.axhspan(B_diss - 3500, B_diss, color='#d6eaf8', alpha=0.60, zorder=0)
axr.axhspan(B_diss, E_hi, color='#d5f5e3', alpha=0.55, zorder=0)
axr.axhline(B_diss, color='#4a7c59', lw=1.4, ls='--', alpha=0.80, zorder=3)

E_cont = np.linspace(B_diss, E_hi - 500, 1000)
nu_peak = B_diss + 3500
sig_cont = 2800.
sigma_cont = np.exp(-0.5 * ((E_cont - nu_peak) / sig_cont)**2)
axr.fill_betweenx(E_cont, 0, sigma_cont * 0.90, alpha=0.30, color='#4a7c59', zorder=3)
axr.plot(sigma_cont * 0.90, E_cont, color='#4a7c59', lw=2.5, zorder=5)

for v in range(22):
    Ev = E_vib_B(v)
    if Ev >= B_diss:
        break
    fc = 0.22 * np.exp(-0.5 * ((Ev - (B_min + 4200)) / 2600)**2) + 0.04
    fc = np.clip(fc, 0.01, 1.0)
    axr.hlines(Ev, 0, fc, color='#8b1a1a', lw=2.0, alpha=0.78, zorder=4)
    axr.plot(fc, Ev, 'o', ms=3.5, color='#8b1a1a', alpha=0.85, zorder=5)

axr.text(0.50, 0.30, 'Bound–bound\n(discrete)', ha='center', va='center', transform=axr.transAxes, 
         fontsize=9.5, color='#8b1a1a', fontweight='bold',
         bbox=dict(fc='white', ec='#8b1a1a', lw=0.9, boxstyle='round,pad=0.4', alpha=0.97))
axr.text(0.50, 0.76, 'Bound–free\n(continuum)', ha='center', va='center', transform=axr.transAxes, 
         fontsize=9.5, color='#4a7c59', fontweight='bold',
         bbox=dict(fc='white', ec='#4a7c59', lw=0.9, boxstyle='round,pad=0.4', alpha=0.97))

axr.set_xlabel(r'$\sigma(\tilde{\nu})\ \mathrm{(arb.)}$', fontsize=12)
axr.set_xlim(-0.05, 1.35)
axr.set_xticks([0, 0.5, 1.0])
axr.tick_params(which='major', direction='in', length=6, right=True, labelleft=False, left=False)
axr.tick_params(which='minor', direction='in', length=3, right=True)
axr.xaxis.set_minor_locator(AutoMinorLocator())
axr.text(0.10, 0.975, '(b)', transform=axr.transAxes, fontsize=13, fontweight='bold', va='top')

plt.tight_layout()
plt.savefig("/storage/home/hcoda1/9/pagrawala6/r-js130-0/thesisplots/newfolder2/finalPEC.svg", format="svg",
            dpi=400, bbox_inches="tight", facecolor="white")
plt.show()
```

```python
"""
Plot 3: The ∆re Counterfactual
Two-column comparison: what if the B state had the same equilibrium
geometry as X (∆re=0) vs. the actual case (∆re=0.363 Å).
Directly demonstrates why the continuum dominates for O2 SR system.
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
import matplotlib.patches as mpatches

plt.rcParams.update({
    "font.family": "serif",
    "font.size": 11,
    "axes.linewidth": 1.2,
    "xtick.direction": "in",
    "ytick.direction": "in",
})

# ─── Physical parameters ──────────────────────────────────────────────────────
re_X = 1.207
De_X, a_X = 41256., 2.680
re_B_real = 1.604
De_B, a_B = 8110., 1.800
B_min_real = 48990.
B_diss_real = 57100.

def V_X(R):
    return De_X * (1 - np.exp(-a_X * (R - re_X)))**2

def V_B_hypothetical(R, re_B, B_min):
    """Morse for B state with specified re and B_min."""
    return B_min + De_B * (1 - np.exp(-a_B * (R - re_B)))**2

sigma_X = 0.055   # HO wavefunction spread for X state
sigma_B = 0.083   # HO spread for B state

def ho_psi(R_arr, n, re, sigma):
    x = (R_arr - re) / sigma
    if n == 0:
        return np.pi**(-0.25) * np.exp(-0.5 * x**2) / np.sqrt(sigma)
    psi_prev = np.pi**(-0.25) * np.exp(-0.5 * x**2) / np.sqrt(sigma)
    psi_curr = np.sqrt(2.0) * x * psi_prev
    for k in range(1, n):
        psi_next = (np.sqrt(2./(k+1)) * x * psi_curr
                    - np.sqrt(float(k)/(k+1)) * psi_prev)
        psi_prev = psi_curr
        psi_curr = psi_next
    return psi_curr

def E_vib_X(v):
    return 1580.*(v + 0.5) - 12.0*(v + 0.5)**2

def E_vib_B(v, B_min):
    return B_min + 700.*(v + 0.5) - 15.1*(v + 0.5)**2

R = np.linspace(0.82, 4.0, 5000)
Vx = V_X(R)

# ─── Two scenarios ────────────────────────────────────────────────────────────
scenarios = [
    {
        're_B'     : re_X,           # HYPOTHETICAL: aligned with X
        'B_min'    : 48990.,
        'B_diss'   : B_diss_real,
        'label'    : r'Hypothetical: $\Delta r_e = 0$',
        'sublabel' : r'($r_e^\prime = r_e^{\prime\prime} = 1.207\ \AA$)',
        'col_B'    : '#6b3a8b',
        'col_cont' : '#9c66c5',
        'col_disc' : '#c8a0e8',
        'col_spec_cont' : '#7b5aa0',
        'col_spec_disc' : '#c8a0e8',
        'panel_letter': '(a)',
    },
    {
        're_B'     : re_B_real,      # ACTUAL
        'B_min'    : B_min_real,
        'B_diss'   : B_diss_real,
        'label'    : r'Actual: $\Delta r_e = 0.363\ \AA$',
        'sublabel' : r'($r_e^\prime = 1.604\ \AA$,  $r_e^{\prime\prime} = 1.207\ \AA$)',
        'col_B'    : '#8b1a1a',
        'col_cont' : '#b84040',
        'col_disc' : '#d48888',
        'col_spec_cont' : '#2b6bbd',
        'col_spec_disc' : '#8b1a1a',
        'panel_letter': '(b)',
    },
]

# ─── Figure layout ────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(14, 8.5))
# 2 scenario columns, each with a top PEC panel and a bottom spectrum panel
outer = gridspec.GridSpec(1, 2, wspace=0.06)

E_lo, E_hi = -2500, 63000
WX = 3600
WB = 3200

axs = []
for si, sc in enumerate(scenarios):
    inner = gridspec.GridSpecFromSubplotSpec(
        2, 1, subplot_spec=outer[si],
        height_ratios=[2.2, 1.0], hspace=0.12
    )
    ax_pec  = fig.add_subplot(inner[0])
    ax_spec = fig.add_subplot(inner[1])
    axs.append((ax_pec, ax_spec))

    re_B   = sc['re_B']
    B_min  = sc['B_min']
    B_diss = sc['B_diss']
    Vb = V_B_hypothetical(R, re_B, B_min)

    # ── PEC curves ────────────────────────────────────────────────────────────
    ax_pec.plot(R[Vx < 43000], Vx[Vx < 43000],
                color='#1a3a6b', lw=2.2, zorder=5)
    mask_B = (Vb < E_hi) & (R > re_B * 0.70) & (R < 3.8)
    ax_pec.plot(R[mask_B], Vb[mask_B], color=sc['col_B'], lw=2.2, zorder=5)

    # Dissociation limit
    ax_pec.axhline(B_diss, color=sc['col_B'], lw=1.0, ls='--', alpha=0.5)

    # ── X state wavefunction v''=0 ────────────────────────────────────────────
    Ev_X0 = E_vib_X(0)
    psi_X0 = ho_psi(R, 0, re_X, sigma_X)
    psi_disp = psi_X0 * WX + Ev_X0
    mask_wX = (Vx < Ev_X0 + 4000) & (R > 0.88) & (R < 2.4)
    ax_pec.plot(R[mask_wX], psi_disp[mask_wX],
                color='#2b6bbd', lw=2.0, zorder=6)
    ax_pec.axhline(Ev_X0, color='#2b6bbd', lw=0.5, ls='-', alpha=0.35,
                   xmin=0.02, xmax=0.35)
    ax_pec.text(0.84, Ev_X0 + 100, r"$v''=0$", ha='right', fontsize=9,
                color='#2b6bbd')

    # ── B state wavefunctions ─────────────────────────────────────────────────
    for v in [0, 5, 10, 15]:
        Ev_B = E_vib_B(v, B_min)
        if Ev_B > B_diss + 2000:
            continue
        psi_Bv = ho_psi(R, v, re_B, sigma_B)
        psi_Bdisp = psi_Bv * WB + Ev_B
        mask_wB = (Vb < Ev_B + 4000) & (R > re_B * 0.68) & (R < 3.5)
        shade = 0.30 + 0.08 * v
        ax_pec.plot(R[mask_wB], psi_Bdisp[mask_wB],
                    color=sc['col_B'], lw=1.1, alpha=min(shade, 0.7), zorder=4)
        ax_pec.axhline(Ev_B, color=sc['col_B'], lw=0.4, ls='-', alpha=0.25,
                       xmin=0.0, xmax=0.9)

    # ── Continuum wavefunction ────────────────────────────────────────────────
    E_cont = B_diss + 2500
    R_cont = np.linspace(re_B * 0.70, 3.7, 1000)
    Vc = V_B_hypothetical(R_cont, re_B, B_min)
    allowed = Vc < E_cont
    kR = np.sqrt(np.maximum(E_cont - Vc, 0.0))
    kR_n = kR / (kR.max() + 1e-10)
    phase = np.cumsum(kR_n) * 0.40
    env = np.exp(-0.35 * (R_cont - re_B * 0.70))
    env = np.clip(env, 0, 1)
    psi_c = np.sin(phase) * env * allowed * 0.55
    ax_pec.plot(R_cont, psi_c * WB * 0.85 + E_cont,
                color='#226633', lw=1.4, alpha=0.85, zorder=4)

    # ── FC window (vertical column from X v''=0 TPs) ─────────────────────────
    sqrt_E = np.sqrt(E_vib_X(0) / De_X)
    R_tp_i = re_X - (1./a_X) * np.log(1. + sqrt_E)
    R_tp_o = re_X - (1./a_X) * np.log(1. - sqrt_E)
    ax_pec.axvspan(R_tp_i - 0.01, R_tp_o + 0.01, alpha=0.12,
                   color='gold', zorder=1)

    # Project FC window to B state — energy range
    E_B_at_inner = V_B_hypothetical(R_tp_i, re_B, B_min)
    E_B_at_outer = V_B_hypothetical(R_tp_o, re_B, B_min)
    E_B_at_re    = V_B_hypothetical(re_X, re_B, B_min)

    # Vertical arrow at re_X
    ax_pec.annotate('', xy=(re_X, E_B_at_re + 500),
                    xytext=(re_X, E_vib_X(0) + 200),
                    arrowprops=dict(arrowstyle='->', color='#226633',
                                    lw=2.2, mutation_scale=14))

    # Shaded excitation target region on B PEC
    R_proj = R[(R >= R_tp_i) & (R <= R_tp_o)]
    E_B_proj = V_B_hypothetical(R_proj, re_B, B_min)
    if len(R_proj) > 1:
        ax_pec.fill_between(R_proj, E_B_proj - 500, E_B_proj + 500,
                            color='gold', alpha=0.40, zorder=8)

    # Annotation: continuum vs discrete
    frac_above = 1.0 if E_B_at_re > B_diss else 0.0
    if si == 0:
        msg = "FC window\nlargely in\n" + r"$\it{discrete}$" + " region"
        msg_col = sc['col_disc']
    else:
        msg = "FC window\nmostly in\n" + r"$\it{continuum}$"
        msg_col = '#226633'
    ax_pec.text(re_X + 0.08, (E_vib_X(0) + E_B_at_re) * 0.5 + 2000,
                msg, fontsize=8.5, color=msg_col, va='center', ha='left',
                bbox=dict(fc='white', ec=msg_col, lw=0.7,
                          boxstyle='round,pad=0.25', alpha=0.92))

    # ── ∆re annotation ────────────────────────────────────────────────────────
    y_dre = B_min - 3000
    ax_pec.annotate('', xy=(re_B, y_dre), xytext=(re_X, y_dre),
                    arrowprops=dict(arrowstyle='<->', color='black', lw=1.4,
                                    mutation_scale=10))
    delta_str = (r'$\Delta r_e = 0$' if si == 0
                 else r'$\Delta r_e = 0.363\ \AA$')
    ax_pec.text((re_X + re_B) / 2, y_dre - 1400, delta_str,
                ha='center', fontsize=9.5, fontstyle='italic')
    if abs(re_B - re_X) > 0.01:
        ax_pec.plot([re_X, re_X], [E_lo, y_dre], color='#1a3a6b',
                    lw=0.6, ls=':', alpha=0.3, zorder=2)
        ax_pec.plot([re_B, re_B], [B_min - 100, y_dre], color=sc['col_B'],
                    lw=0.6, ls=':', alpha=0.3, zorder=2)

    # Labels and ticks
    ax_pec.set_xlim(0.82, 3.8)
    ax_pec.set_ylim(E_lo, E_hi)
    ax_pec.tick_params(which='major', direction='in', length=5, top=True, right=True)
    ax_pec.tick_params(which='minor', direction='in', length=3, top=True, right=True)
    ax_pec.xaxis.set_minor_locator(AutoMinorLocator())
    ax_pec.yaxis.set_minor_locator(AutoMinorLocator())
    ax_pec.yaxis.set_major_locator(MultipleLocator(10000))
    if si == 0:
        ax_pec.set_ylabel(r'Energy $(\mathrm{cm}^{-1})$', fontsize=11)
    else:
        ax_pec.set_yticklabels([])

    # State labels
    ax_pec.text(3.0, 7000, r'X $^3\Sigma_g^-$', color='#1a3a6b',
                fontsize=10.5, fontweight='bold')
    ax_pec.text(2.9, B_min + 1800, r'B $^3\Sigma_u^-$', color=sc['col_B'],
                fontsize=10.5, fontweight='bold')
    ax_pec.text(0.012, 0.975, sc['panel_letter'],
                transform=ax_pec.transAxes, fontsize=12, fontweight='bold', va='top')

    # Title per panel
    ax_pec.set_title(sc['label'] + '\n' + sc['sublabel'], fontsize=11, pad=6)

    # ── Spectrum panel (qualitative stick + continuum) ─────────────────────────
    # x = transition wavenumber (cm⁻¹), y = cross-section
    ax_spec.set_xlim(47000, 64000)

    # Discrete bands (v' below B_diss)
    disc_contributions = []
    for v in range(25):
        Ev_B = E_vib_B(v, B_min)
        if Ev_B >= B_diss:
            break
        nu_trans = Ev_B - E_vib_X(0)
        # Qualitative FC factor
        if si == 0:
            # Aligned: strong near 0-0, falls off quickly
            fc = 0.85 * np.exp(-0.5 * ((nu_trans - 49800) / 800)**2) + 0.05
        else:
            # Displaced: weak discrete, peaks near v'~15 (high-v)
            fc = 0.15 * np.exp(-0.5 * ((nu_trans - 55500) / 1400)**2) + 0.02
        disc_contributions.append((nu_trans, fc))
        ax_spec.plot([nu_trans, nu_trans], [0, fc],
                     color=sc['col_spec_disc'], lw=1.8, alpha=0.75, zorder=5)
        ax_spec.plot(nu_trans, fc, 'o', ms=3, color=sc['col_spec_disc'], alpha=0.75)

    # Continuum envelope
    nu_arr = np.linspace(B_diss - E_vib_X(0), 64000 - E_vib_X(0), 500)
    E_B_arr = nu_arr + E_vib_X(0)
    if si == 0:
        # Continuum exists but weak (FC window didn't project into it much)
        cont_peak_E = B_min + De_B * 0.1
        sigma_cont = 1500.
        cont_amp = 0.18
    else:
        # Strong continuum
        cont_peak_E = B_diss + 2000
        sigma_cont = 3000.
        cont_amp = 0.90
    sigma_c = cont_amp * np.exp(-0.5*((E_B_arr - cont_peak_E) / sigma_cont)**2)
    sigma_c = np.clip(sigma_c, 0, None)
    nu_plot = nu_arr + E_vib_X(0) - B_diss + B_diss
    # Map to transition wavenumber
    nu_trans_arr = nu_arr
    ax_spec.fill_between(nu_trans_arr + E_vib_X(0), 0, sigma_c,
                         color=sc['col_spec_cont'], alpha=0.25, zorder=3)
    ax_spec.plot(nu_trans_arr + E_vib_X(0), sigma_c,
                 color=sc['col_spec_cont'], lw=2.0, zorder=4)

    # Continuum boundary
    ax_spec.axvline(B_diss - E_vib_X(0) + E_vib_X(0), color='gray',
                    lw=0.9, ls='--', alpha=0.5, zorder=2)
    ax_spec.text(B_diss + 100, 0.95,
                 r'$D_e(\mathrm{B})$', color='gray', fontsize=8.5,
                 va='top', transform=ax_spec.get_xaxis_transform())

    ax_spec.set_xlabel(r'Transition wavenumber $\tilde{\nu}\ (\mathrm{cm}^{-1})$',
                       fontsize=10)
    ax_spec.set_ylim(0, 1.05)
    ax_spec.set_xlim(48800, 63000)
    if si == 0:
        ax_spec.set_ylabel(r'$\sigma\ (\mathrm{arb.})$', fontsize=10)
    else:
        ax_spec.set_yticklabels([])
    ax_spec.tick_params(which='major', direction='in', length=5, top=True, right=True)
    ax_spec.tick_params(which='minor', direction='in', length=3, top=True, right=True)
    ax_spec.xaxis.set_minor_locator(AutoMinorLocator())
    ax_spec.yaxis.set_minor_locator(AutoMinorLocator())

    # Outcome annotation
    if si == 0:
        outcome = r"$0\!-\!0$ band dominates"
        outcome_col = sc['col_disc']
    else:
        outcome = "Continuum\ndominates (>99%)"
        outcome_col = sc['col_spec_cont']
    ax_spec.text(0.97, 0.92, outcome, ha='right', va='top',
                 transform=ax_spec.transAxes, fontsize=9.5,
                 color=outcome_col, fontweight='bold',
                 bbox=dict(fc='white', ec=outcome_col, lw=0.7,
                           boxstyle='round,pad=0.3', alpha=0.9))

fig.suptitle(
    r'The $\Delta r_e$ Counterfactual: How Equilibrium Displacement Controls'
    ' the Oscillator Strength Budget',
    fontsize=12.5, y=1.01, fontweight='bold'
)

# plt.savefig('/mnt/user-data/outputs/plot3_delta_re_counterfactual.png',
#             dpi=160, bbox_inches='tight')
# print("Saved plot3_delta_re_counterfactual.png")
```

```python
"""
Plot C — Thermal Decomposition (standalone, polished)
Top panel  : σ(ν̃, T) for T = 300, 1000, 3000 K — three thermal curves
Bottom-left : stacked-area decomposition at T = 300 K
Bottom-right: stacked-area decomposition at T = 3000 K
Each bottom panel has a Boltzmann population inset.
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
from matplotlib.colors import Normalize
import matplotlib.cm as cm
from scipy.ndimage import gaussian_filter1d

# ── global style ──────────────────────────────────────────────────────────────
plt.rcParams.update({
    "font.family":        "serif",
    "font.size":          12,
    "axes.labelsize":     12,
    "axes.titlesize":     12,
    "xtick.labelsize":    10.5,
    "ytick.labelsize":    10.5,
    "axes.linewidth":     1.4,
    "xtick.direction":    "in",
    "ytick.direction":    "in",
    "xtick.major.size":   5,
    "ytick.major.size":   5,
    "xtick.minor.size":   2.8,
    "ytick.minor.size":   2.8,
    "legend.fontsize":    10.5,
    "legend.framealpha":  0.92,
})

# ── physics ───────────────────────────────────────────────────────────────────
re_X, De_X, a_X = 1.207, 41256., 2.680
re_B, De_B, a_B = 1.604,  8110., 1.800
B_min  = 48990.
B_diss = 57100.
k_B_cm = 0.6950356

def V_B(R): return B_min + De_B * (1 - np.exp(-a_B * (R - re_B)))**2
def E_vib_X(v): return 1580.*(v + 0.5) - 12.0*(v + 0.5)**2
def TDM(R):  return 1.8 * np.exp(-0.5 * ((R - 1.26) / 0.22)**2)

def ho_psi(R_arr, n, re, sigma):
    x = (R_arr - re) / sigma
    p = np.pi**(-0.25) * np.exp(-0.5 * x**2) / np.sqrt(sigma)
    if n == 0: return p
    c = np.sqrt(2.0) * x * p
    for k in range(1, n):
        tmp = np.sqrt(2./(k+1)) * x * c - np.sqrt(float(k)/(k+1)) * p
        p = c; c = tmp
    return c

def boltzmann_pop(n_v, T):
    E = np.array([E_vib_X(v) - E_vib_X(0) for v in range(n_v)])
    w = np.exp(-E / (k_B_cm * T))
    return w / w.sum()

# ── compute σ(ν̃, v'') for v''=0..15 ─────────────────────────────────────────
R  = np.linspace(0.85, 5.0, 6000)
Vb = V_B(R)
tdm_R = TDM(R)

nu_lo, nu_hi = 57100., 66500.
n_nu  = 200
n_vpp = 16
nu_arr = np.linspace(nu_lo, nu_hi, n_nu)
dnu    = nu_arr[1] - nu_arr[0]

print("Computing σ(ν̃, v'') …")
sigma_smooth = np.zeros((n_vpp, n_nu))
for iv in range(n_vpp):
    sigma_eff = 0.055 * np.sqrt(2*iv + 1) * 0.60 + 0.055 * 0.40
    psi_v = ho_psi(R, iv, re_X, sigma_eff)
    g = tdm_R * psi_v
    raw = np.zeros(n_nu)
    for inu, nu in enumerate(nu_arr):
        Ec = nu + E_vib_X(iv)
        if Ec <= B_diss + 50: continue
        kR = np.sqrt(np.maximum(Ec - Vb, 0.0))
        k_s = np.where(kR > 0, kR, 1e-8)
        env = np.clip(np.exp(-0.12*(R - 1.0)), 0.02, 1.0)
        raw[inu] = nu * np.trapz(g**2 * env**2 / k_s * (kR > 0), R)
    sigma_smooth[iv] = gaussian_filter1d(raw, sigma=60./dnu)
print("Done.")

# thermal averages
temps = [300, 1000, 3000]
sigma_T = {}
for T in temps:
    pop = boltzmann_pop(n_vpp, T)
    sigma_T[T] = sum(pop[iv] * sigma_smooth[iv] for iv in range(n_vpp))

sig_300_max = sigma_T[300].max()

# colour map for v'' layers
cmap_v = cm.plasma_r
norm_v = Normalize(0, n_vpp - 1)

# ── figure ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(13, 9.5))
fig.patch.set_facecolor("white")

gs = gridspec.GridSpec(
    2, 2,
    figure=fig,
    hspace=0.38, wspace=0.10,
    height_ratios=[1.0, 1.05],
    top=0.91, bottom=0.08, left=0.08, right=0.96,
)
ax_top = fig.add_subplot(gs[0, :])          # full-width top
ax_lo  = fig.add_subplot(gs[1, 0])          # bottom-left  300 K
ax_hi  = fig.add_subplot(gs[1, 1],          # bottom-right 3000 K
                          sharey=ax_lo)

T_cols   = ["#1a3a6b", "#7b3f9e", "#b22222"]
T_labels = ["300 K", "1000 K", "3000 K"]

# ════════════════════════════════════════════════════════════════════════════
# Top panel — three thermal curves
# ════════════════════════════════════════════════════════════════════════════
ax = ax_top
for T, col, lbl in zip(temps, T_cols, T_labels):
    ax.plot(nu_arr, sigma_T[T] / sig_300_max,
            color=col, lw=2.2, label=lbl, zorder=4)

# shaded band between 300 K and 3000 K
ax.fill_between(nu_arr,
                sigma_T[300]  / sig_300_max,
                sigma_T[3000] / sig_300_max,
                alpha=0.13, color="#b22222", zorder=2,
                label="Temperature variation (300 → 3000 K)")

# annotation: spectral shift
ax.annotate("",
            xy=(59400, sigma_T[3000][np.argmin(np.abs(nu_arr-59400))]/sig_300_max + 0.04),
            xytext=(59400, sigma_T[300][np.argmin(np.abs(nu_arr-59400))]/sig_300_max - 0.04),
            arrowprops=dict(arrowstyle="<->", color="#b22222",
                            lw=1.6, mutation_scale=12), zorder=6)
# ax.text(59700, 0.70,
#         r"$\sigma$ changes with $T$" + "\n(state-resolved effect)",
#         fontsize=10.5, color="#b22222", va="center",
#         bbox=dict(fc="white", ec="#b22222", lw=0.8,
#                   boxstyle="round,pad=0.25", alpha=0.93))

ax.set_xlim(nu_lo, nu_hi)
ax.set_ylim(-0.03, 1.15)
ax.set_xticklabels([])
ax.set_ylabel(r"$\sigma(\tilde{\nu},\,T)$  [norm. to 300 K peak]", fontsize=12)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which="major", direction="in", length=5,  top=True, right=True)
ax.tick_params(which="minor", direction="in", length=2.8, top=True, right=True)
ax.legend(loc="upper right", fontsize=10.5, handlelength=1.6,
          borderpad=0.5, ncol=2)
ax.text(0.008, 0.965, "(a)", transform=ax.transAxes,
        fontsize=13, fontweight="bold", va="top")
ax.set_title(
    r"O$_2$ Schumann–Runge Continuum: Thermal Average vs. State-Resolved Decomposition",
    fontsize=12.5, pad=7, fontweight="bold",
)

# wavelength axis on top
ax_t2 = ax.twiny()
nu_tix = np.array([57500, 59000, 60500, 62000, 63500, 65000])
ax_t2.set_xlim(nu_lo, nu_hi)
ax_t2.set_xticks(nu_tix)
ax_t2.set_xticklabels([f"{1e7/n:.0f}" for n in nu_tix], fontsize=10)
ax_t2.set_xlabel("Wavelength (nm)", fontsize=11, labelpad=6)
ax_t2.tick_params(direction="in", length=5)

# ════════════════════════════════════════════════════════════════════════════
# Bottom panels — stacked-area decomposition
# ════════════════════════════════════════════════════════════════════════════
def draw_decomp(ax, T, panel_letter, title_str):
    pop = boltzmann_pop(n_vpp, T)

    # stacked fill — build from v''=n_vpp-1 downward so v''=0 is on top
    running = np.zeros(n_nu)
    for iv in range(n_vpp - 1, -1, -1):
        contrib = pop[iv] * sigma_smooth[iv] / sig_300_max
        col = cmap_v(norm_v(iv))
        ax.fill_between(nu_arr, running, running + contrib,
                        color=col, alpha=0.82, zorder=2)
        running += contrib

    # thermal total overlaid
    ax.plot(nu_arr, sigma_T[T] / sig_300_max,
            color="black", lw=2.0, zorder=5, label="Thermal total")

    ax.set_xlim(nu_lo, nu_hi)
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(which="major", direction="in", length=5, top=True, right=True)
    ax.tick_params(which="minor", direction="in", length=2.8, top=True, right=True)
    ax.set_xlabel(r"$\tilde{\nu}$ (cm$^{-1}$)", fontsize=12)
    ax.set_title(title_str, fontsize=11.5, pad=5)
    ax.text(0.010, 0.965, panel_letter, transform=ax.transAxes,
            fontsize=13, fontweight="bold", va="top")

    # ── Boltzmann population inset ────────────────────────────────────────
    ax_ins = ax.inset_axes([0.695, 0.36, 0.285, 0.57])
    for iv in range(min(9, n_vpp)):
        col = cmap_v(norm_v(iv))
        ax_ins.barh(iv, pop[iv] * 100., height=0.70,
                    color=col, alpha=0.88, edgecolor="white", lw=0.4)
    ax_ins.set_yticks(range(min(9, n_vpp)))
    ax_ins.set_yticklabels(
        [f"$v''={v}$" for v in range(min(9, n_vpp))], fontsize=6.8)
    ax_ins.tick_params(labelsize=7.5, direction="in", length=2.5)
    ax_ins.set_xlabel("Pop. (%)", fontsize=7.5)
    ax_ins.set_title("Boltzmann\npopulation", fontsize=7.5, pad=2,
                     linespacing=1.3)
    ax_ins.spines["top"].set_visible(False)
    ax_ins.spines["right"].set_visible(False)
    for spine in ax_ins.spines.values():
        spine.set_linewidth(0.8)

draw_decomp(ax_lo, 300,
            "(b)",
            r"$T = 300$ K — $\sigma(\tilde{\nu})$ dominated by $v''=0$")
draw_decomp(ax_hi, 3000,
            "(c)",
            r"$T = 3000$ K — higher $v''$ layers contribute")

ax_lo.set_ylabel(r"$\sigma$  [norm.]", fontsize=12)
ax_hi.set_yticklabels([])
ax_hi.legend(loc="upper left", fontsize=10.5, handlelength=1.4,
             borderpad=0.5)

# shared y-axis limits
ax_lo.set_ylim(-0.02, 1.12)

# ── colourbar for v'' layers ──────────────────────────────────────────────────
sm = cm.ScalarMappable(cmap="plasma_r", norm=norm_v)
sm.set_array([])
cbar = fig.colorbar(sm, ax=[ax_lo, ax_hi],
                     fraction=0.028, pad=0.018, aspect=22)
cbar.set_label(r"Vibrational level $v''$", fontsize=11.5)
cbar.set_ticks([0, 5, 10, 15])
cbar.ax.tick_params(labelsize=10)

plt.savefig("/storage/home/hcoda1/9/pagrawala6/r-js130-0/thesisplots/newfolder2/continuum_temp.svg", format="svg",
            dpi=400, bbox_inches="tight", facecolor="white")
# plt.close()
print("Saved plotC_thermal_decomp_final.svg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

# --- Professional Plotting Configuration ---
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 11,
    "axes.linewidth": 1.2,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 5,
    "ytick.major.size": 5,
    "xtick.minor.size": 3,
    "ytick.minor.size": 3,
    "legend.frameon": False,
    "lines.linewidth": 1.5,
    "savefig.dpi": 300,
    "savefig.bbox": 'tight'
})

def morse_potential(r, de, re, a):
    """Calculates the Morse potential energy for a diatomic system."""
    return de * (1 - np.exp(-a * (r - re)))**2

def bound_wavefunction(r, re, width):
    """Qualitative approximation of a v=0 ground state (Gaussian-like)."""
    return np.exp(-((r - re)**2) / (2 * width**2))

def continuum_wavefunction(r, r_tp, k):
    """
    Qualitative approximation of a continuum state.
    Exponentially decays before the turning point (r_tp) and 
    oscillates with constant amplitude afterward.
    """
    psi = np.zeros_like(r)
    mask_before = r < r_tp
    mask_after = r >= r_tp
    
    # Exponential tunneling into the barrier
    psi[mask_before] = np.exp(12 * (r[mask_before] - r_tp))
    # Oscillatory wave in the unbound region
    # The 0.8 power slightly chirps the frequency for visual depth
    psi[mask_after] = np.cos(k * (r[mask_after] - r_tp)**0.8)
    return psi

def generate_spectroscopy_plot():
    # 1. Model Parameters
    de = 100.0   # Dissociation energy (well depth)
    re = 2.0     # Equilibrium internuclear distance
    a = 1.2      # Well width parameter
    
    r = np.linspace(1.2, 6.5, 1200)
    v_pot = morse_potential(r, de, re, a)
    
    # 2. State Definitions
    # Bound State (v=0)
    e_bound = 12.0
    psi_b = bound_wavefunction(r, re, width=0.18)
    psi_b_scaled = psi_b * 18.0  # Scaled to be visible on the energy axis
    
    # Continuum State (E > De)
    e_cont = de + 25.0
    # Approximate inner turning point for the continuum energy
    r_tp_cont = re - np.log(1 + np.sqrt(e_cont/de)) / a 
    psi_c = continuum_wavefunction(r, r_tp_cont, k=28)
    psi_c_scaled = psi_c * 12.0 # Scaled for clarity
    
    # 3. Visualization logic
    fig, ax = plt.subplots(figsize=(10, 7))
    
    # Plot Potential Energy Curve (PEC)
    ax.plot(r, v_pot, color='black', lw=3.0, zorder=2,)
    
    # Plot Bound Wavefunction
    # fill_between highlights the probability density region
    ax.fill_between(r, e_bound, e_bound + psi_b_scaled, color='#2b6bbd', alpha=0.35, zorder=3)
    ax.plot(r, e_bound + psi_b_scaled, color='#2b6bbd', lw=2.2, label='Bound State ($\psi_{v=0}$)', zorder=4)
    ax.axhline(e_bound, color='#2b6bbd', lw=1.2, ls='--', alpha=0.4)
    
    # Plot Continuum Wavefunction
    ax.fill_between(r, e_cont, e_cont + psi_c_scaled, color='#4a7c59', alpha=0.35, zorder=3)
    ax.plot(r, e_cont + psi_c_scaled, color='#4a7c59', lw=2.2, label='Continuum State ($E > D_e$)', zorder=4)
    ax.axhline(e_cont, color='#4a7c59', lw=1.2, ls='--', alpha=0.4)
    
    # Highlight Dissociation Limit
    ax.axhline(de, color='#8b1a1a', lw=1.5, ls=':', alpha=0.7)
    ax.text(r[-1], de + 2, 'Dissociation Limit ($D_e$)', ha='right', va='bottom', 
            fontsize=11, style='italic', color='#8b1a1a')
    
    # 4. Refinement and Styling
    ax.set_xlabel('Internuclear Distance ($r$)', fontsize=13)
    ax.set_ylabel('Energy ($E$)', fontsize=13)
    ax.set_xlim(r[0], r[-1])
    ax.set_ylim(-15, de + 70)
    
    # Remove top/right spines for a cleaner aesthetic
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Legend placement
    ax.legend(loc='upper right', frameon=False, fontsize=11)
    
    # Tick adjustments
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    
    # Annotations to explain the physics
    ax.text(re + 0.3, e_bound + 5, "Localized:\nDecays to 0", color='#1a3a6b', 
            fontsize=10, fontweight='bold', ha='left')
    ax.text(4.5, e_cont + 15, "Delocalized:\nOscillates to $\infty$", color='#226633', 
            fontsize=10, fontweight='bold', ha='center')
    
    plt.tight_layout()
    plt.savefig('/storage/home/hcoda1/9/pagrawala6/r-js130-0/thesisplots/plot_spectroscopy_concept.pdf', dpi=300)
    plt.show()

if __name__ == "__main__":
    generate_spectroscopy_plot()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from matplotlib.ticker import AutoMinorLocator
import matplotlib.gridspec as gridspec

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
})

# ─────────────────────────────────────────────────────────────────
# Physical parameters (cm⁻¹, approximate but representative)
# X³Σg⁻: odd N only (nuclear spin statistics, I=0 bosons)
# B³Σu⁺: even N only
# F1: J=N+1, F2: J=N, F3: J=N-1
# Fine structure splitting O2 X state: λ≈59.5 cm⁻¹, γ≈-0.25 cm⁻¹
# B_X ≈ 1.4377, B_B ≈ 0.8190 cm⁻¹
# ─────────────────────────────────────────────────────────────────

B_X = 1.4377
lam_X = 59.501   # spin-spin parameter
gam_X = -0.253   # spin-rotation

B_B = 0.8190
lam_B = 0.0      # B state is purely repulsive above dissoc. for continuum
                  # but we use approximate bound-state values for illustration
gam_B = 0.0

def F_levels_3Sigma(N, B, lam, gam):
    """
    Rotational energies for ³Σ state (Hund's case b), three F components.
    Formulae: Herzberg Vol. I, or Brown & Carrington.
    F1: J=N+1, F2: J=N, F3: J=N-1
    """
    EN = B * N * (N + 1)
    if N == 0:
        # Only F1 exists (J=1), F2 and F3 don't exist for N=0
        E_F1 = EN + (2 * lam) / 3 + gam * (N + 1)
        return {'F1': E_F1, 'F2': None, 'F3': None}
    D_N = np.sqrt((N + 0.5)**2 + (lam / B) * (1 + (lam / B) / (4 * (N + 0.5)**2)))
    # Simplified approximation for illustration
    split = lam * 2 / (2 * N + 1) if N > 0 else 0
    E_F2 = EN
    E_F1 = EN + split * (N + 1) / (2 * N + 1) + gam * N
    E_F3 = EN - split * N / (2 * N + 1) - gam * (N + 1)
    return {'F1': E_F1, 'F2': E_F2, 'F3': E_F3}

# Build energy level dictionary for diagram
# X state: N = 1, 3 (odd); B state: N = 2, 4 (even)
# Use scaled energies for visual clarity in the diagram
# Scale: divide by 100 to get diagram units

def diagram_scale(E): return E / 80.0

X_levels = {}
for N in [1, 3]:
    X_levels[N] = F_levels_3Sigma(N, B_X, lam_X, gam_X)

B_levels = {}
T_e = 49794  # approximate T_e for B state in cm⁻¹; offset for diagram
B_offset = 400  # visual offset for diagram units
for N in [2, 4]:
    raw = F_levels_3Sigma(N, B_B, 1.5, 0.0)  # use small splitting for B state
    B_levels[N] = {k: (v + B_offset if v is not None else None) for k, v in raw.items()}

# ─────────────────────────────────────────────────────────────────
# Figure layout: 2 columns
# Left: Grotrian diagram (full height)
# Right: top = fine-structure zoom, bottom = synthetic spectrum
# ─────────────────────────────────────────────────────────────────

fig = plt.figure(figsize=(14, 9))
gs = gridspec.GridSpec(2, 2, width_ratios=[1.1, 1.0], height_ratios=[1.9, 1.0],
                        wspace=0.35, hspace=0.38)

# ─────────────────────────────────────────────────────────────────
# Panel A: Grotrian diagram
# ─────────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[:, 0])
ax1.axis('off')
ax1.set_xlim(0, 1)
ax1.set_ylim(-0.5, 13)
ax1.set_title(r"O$_2$ Schumann-Runge: X $^3\Sigma_g^-$ $\rightarrow$ B $^3\Sigma_u^+$"
              "\nRotational branch structure (12 branches)",
              fontsize=12, pad=8)

# --- Layout parameters ---
x_left   = 0.08   # left edge of levels
x_right  = 0.68   # right edge of levels (before labels)
x_label  = 0.70   # F-label x position

# --- X state levels (N=1, N=3) ---
# Map N to diagram y-baseline
X_base = {1: 0.5, 3: 2.5}

# --- B state levels (N=2, N=4) ---
B_base = {2: 7.0, 4: 9.5}

col_main = {'F1': '#222222', 'F2': '#555555', 'F3': '#888888'}
lw_level = 1.5

def draw_levels(ax, N, base, levels, state_label, side='X'):
    """Draw three F-component lines for a given N."""
    offsets = {'F1': 0.35, 'F2': 0.0, 'F3': -0.35}
    for fi, (key, E) in enumerate(levels.items()):
        if E is None:
            continue
        y = base + offsets[key]
        ax.hlines(y, x_left, x_right, color=col_main[key], linewidth=lw_level, zorder=3)
        # F label on right
        ax.text(x_label, y, f"$F_{3-fi}$", fontsize=9, va='center', color=col_main[key])
        # J label
        J = N + 1 - fi
        ax.text(x_label + 0.12, y, f"J={J}", fontsize=8, va='center', color='gray')

    # N label on left, centred on the triplet
    ax.text(x_left - 0.04, base, f"N={N}", fontsize=10, va='center', ha='right',
            fontweight='bold')

# Draw a dashed "centre of gravity" line for each N (nuclear rotation)
for N, base in X_base.items():
    ax1.hlines(base, x_left - 0.01, x_left, color='black', linewidth=0.5, linestyle='--')
    draw_levels(ax1, N, base, X_levels[N], 'X', side='X')

for N, base in B_base.items():
    ax1.hlines(base, x_left - 0.01, x_left, color='black', linewidth=0.5, linestyle='--')
    draw_levels(ax1, N, base, B_levels[N], 'B', side='B')

# State labels
ax1.text(-0.02, 1.5, r"X $^3\Sigma_g^-$" + "\n(odd N only)", fontsize=10,
         va='center', ha='left', color='navy',
         bbox=dict(fc='aliceblue', ec='navy', lw=0.8, boxstyle='round,pad=0.3'))
ax1.text(-0.02, 8.25, r"B $^3\Sigma_u^+$" + "\n(even N only)", fontsize=10,
         va='center', ha='left', color='darkred',
         bbox=dict(fc='#fff0f0', ec='darkred', lw=0.8, boxstyle='round,pad=0.3'))

# Gap indicator (dissociation continuum region)
ax1.axhspan(4.5, 5.5, color='lightyellow', alpha=0.7, zorder=0)
ax1.text(0.5, 5.0, "— — — Dissociation Limit — — —", ha='center', va='center',
         fontsize=9, color='goldenrod', style='italic')

# ─────────────────────────────────────────────────────────────────
# Draw transitions
# Convention:
#   Main branches (ΔFi=0): colour by branch type R/P
#   Satellite branches (ΔFi≠0): muted/dashed
# Show transitions from N''=1 (X) to N'=2 (B)
# and N''=3 (X) to N'=4 (B)  [R branches]
# and N''=3 (X) to N'=2 (B)  [P branches]
# ─────────────────────────────────────────────────────────────────

F_offsets = {'F1': 0.35, 'F2': 0.0, 'F3': -0.35}
F_keys = ['F1', 'F2', 'F3']

# x-positions for the arrows (spread across x_left to x_right)
arrow_x_main = {'R': [0.15, 0.22, 0.29],   # R1, R2, R3 (ΔF=0)
                 'P': [0.36, 0.43, 0.50]}
arrow_x_sat  = [0.57, 0.60, 0.63]

col_R = '#1a6faf'
col_P = '#b22222'
col_sat = '#7b3f9e'

def draw_arrow(ax, x, y_bottom, y_top, color, lw=1.8, ls='-', head='->'):
    ax.annotate('', xy=(x, y_top), xytext=(x, y_bottom),
                arrowprops=dict(arrowstyle=head, color=color, lw=lw,
                                linestyle=ls, mutation_scale=12))

# --- Main R branches: N''=1 → N'=2 ---
for fi, fkey in enumerate(F_keys):
    y_lower = X_base[1] + F_offsets[fkey]
    y_upper = B_base[2] + F_offsets[fkey]
    draw_arrow(ax1, arrow_x_main['R'][fi], y_lower, y_upper, col_R, lw=2.0)

# Label R branch group
ax1.text(arrow_x_main['R'][1], (X_base[1] + B_base[2]) / 2,
         "R$_1$, R$_2$, R$_3$\n(main)", ha='center', va='center',
         fontsize=8.5, color=col_R,
         bbox=dict(fc='white', ec=col_R, lw=0.6, boxstyle='round,pad=0.2'))

# --- Main P branches: N''=3 → N'=2 ---
for fi, fkey in enumerate(F_keys):
    y_lower = X_base[3] + F_offsets[fkey]
    y_upper = B_base[2] + F_offsets[fkey]
    draw_arrow(ax1, arrow_x_main['P'][fi], y_lower, y_upper, col_P, lw=2.0)

ax1.text(arrow_x_main['P'][1], (X_base[3] + B_base[2]) / 2,
         "P$_1$, P$_2$, P$_3$\n(main)", ha='center', va='center',
         fontsize=8.5, color=col_P,
         bbox=dict(fc='white', ec=col_P, lw=0.6, boxstyle='round,pad=0.2'))

# --- Satellite branches (ΔFi ≠ 0): N''=1 → N'=2, selected ---
# RQ21: R-type ΔN, F2'←F1'', i.e. F2 upper, F1 lower
sat_transitions = [
    ('RQ$_{21}$', B_base[2]+F_offsets['F2'], X_base[1]+F_offsets['F1']),
    ('RQ$_{32}$', B_base[2]+F_offsets['F3'], X_base[1]+F_offsets['F2']),
    (r'$^R$P$_{31}$', B_base[2]+F_offsets['F3'], X_base[1]+F_offsets['F1']),
]
for i, (lbl, y_top, y_bot) in enumerate(sat_transitions):
    xpos = arrow_x_sat[i]
    draw_arrow(ax1, xpos, y_bot, y_top, col_sat, lw=1.2, ls='dashed')

# Satellite label
ax1.text(arrow_x_sat[1], (X_base[1] + B_base[2])/2 + 0.5,
         "RQ$_{21}$, RQ$_{32}$\n$^R$P$_{31}$ (sat.)",
         ha='center', va='center', fontsize=8, color=col_sat,
         bbox=dict(fc='white', ec=col_sat, lw=0.6, boxstyle='round,pad=0.2'))

# P-type satellites: N''=3 → N'=2
sat_P = [
    ('PQ$_{12}$', B_base[2]+F_offsets['F1'], X_base[3]+F_offsets['F2']),
    ('PQ$_{23}$', B_base[2]+F_offsets['F2'], X_base[3]+F_offsets['F3']),
]
sat_xP = [0.54, 0.58]
for i, (lbl, y_top, y_bot) in enumerate(sat_P):
    draw_arrow(ax1, sat_xP[i], y_bot, y_top, col_sat, lw=1.2, ls='dashed')

ax1.text(sat_xP[0]+0.02, (X_base[3]+B_base[2])/2,
         "PQ$_{12}$, PQ$_{23}$\n(sat.)",
         ha='center', va='center', fontsize=8, color=col_sat,
         bbox=dict(fc='white', ec=col_sat, lw=0.6, boxstyle='round,pad=0.2'))

# Legend
from matplotlib.lines import Line2D
legend_items = [
    Line2D([0],[0], color=col_R, lw=2, label='R branch (main, \u0394N=+1)'),
    Line2D([0],[0], color=col_P, lw=2, label='P branch (main, \u0394N=\u22121)'),
    Line2D([0],[0], color=col_sat, lw=1.5, linestyle='--', label='Satellite branches (\u0394F\u2260\u0394N)'),
]
ax1.legend(handles=legend_items, loc='lower right', fontsize=8.5, framealpha=0.9)

# ─────────────────────────────────────────────────────────────────
# Panel B: Fine structure energy diagram (top right)
# Shows the actual computed F1/F2/F3 for X N=1 and B N=2
# in physical units (cm⁻¹), relative to each state's N centre
# ─────────────────────────────────────────────────────────────────
ax2 = fig.add_subplot(gs[0, 1])
ax2.set_title("Fine structure of N''=1 (X) and N'=2 (B)\n"
              r"relative to BN(N+1) [cm$^{-1}$]", fontsize=11)

# Compute exact F levels relative to BN(N+1)
def relative_F(N, B, lam, gam):
    EN = B * N * (N + 1)
    lvls = F_levels_3Sigma(N, B, lam, gam)
    return {k: (v - EN if v is not None else None) for k, v in lvls.items()}

x_rel = relative_F(1, B_X, lam_X, gam_X)
b_rel = relative_F(2, B_B, 1.5, 0.0)

col_fi = {'F1': '#1a6faf', 'F2': '#228b22', 'F3': '#b22222'}
line_width = 80  # half-width of level bar in cm⁻¹ (x-axis)

# X state levels at x-centre = 1, B state at x-centre = 3
for fi, fkey in enumerate(list(x_rel.keys())):
    if x_rel[fkey] is None:
        continue
    J = 1 + 1 - fi  # J = N+1-fi for N=1
    ax2.hlines(x_rel[fkey], 0.7, 1.3, color=col_fi[fkey], linewidth=2.5, zorder=3)
    ax2.text(1.35, x_rel[fkey], f"$F_{3-fi}''$  (J={J})",
             va='center', fontsize=9, color=col_fi[fkey])

for fi, fkey in enumerate(list(b_rel.keys())):
    if b_rel[fkey] is None:
        continue
    J = 2 + 1 - fi
    ax2.hlines(b_rel[fkey] + 300, 2.7, 3.3, color=col_fi[fkey], linewidth=2.5, zorder=3)
    ax2.text(3.35, b_rel[fkey] + 300, f"$F_{3-fi}'$  (J={J})",
             va='center', fontsize=9, color=col_fi[fkey])

# Draw main transitions (ΔF=0) as vertical arrows between the two sets
for fi, fkey in enumerate(F_keys):
    if x_rel[fkey] is None or b_rel[fkey] is None:
        continue
    ax2.annotate('', xy=(2.7, b_rel[fkey] + 300), xytext=(1.3, x_rel[fkey]),
                 arrowprops=dict(arrowstyle='->', color=col_fi[fkey], lw=1.8,
                                 connectionstyle='arc3,rad=0.0'))

ax2.text(2.0, 170, r"Main: $\Delta F_i = 0$", ha='center', fontsize=9, color='black',
         style='italic')

ax2.set_xlim(0.4, 4.5)
ax2.set_ylim(-110, 410)
ax2.set_xticks([1.0, 3.0])
ax2.set_xticklabels(["N''=1\n(X state)", "N'=2\n(B state)"], fontsize=10)
ax2.set_ylabel(r"Energy relative to BN(N+1) [cm$^{-1}$]", fontsize=10)
ax2.tick_params(which='major', direction='in', length=5, top=False, right=True, labelsize=9)
ax2.tick_params(which='minor', direction='in', length=3, top=False, right=True)
ax2.yaxis.set_minor_locator(AutoMinorLocator())
ax2.spines['top'].set_visible(False)

# ─────────────────────────────────────────────────────────────────
# Panel C: Synthetic spectrum (bottom right)
# Qualitative: main branches form two closely-spaced groups (R, P)
# with satellite lines at lower intensity on the wings
# ─────────────────────────────────────────────────────────────────
ax3 = fig.add_subplot(gs[1, 1])

def gaussian_peak(x, ctr, amp, w):
    return amp * np.exp(-((x - ctr)**2) / (2 * w**2))

x = np.linspace(0, 12, 2000)
y = np.zeros_like(x)

# R branch group (higher wavenumber side)
R_centres = [8.0, 8.5, 9.0]
R_amps    = [1.0, 1.15, 0.85]
for c, a in zip(R_centres, R_amps):
    y += gaussian_peak(x, c, a, 0.08)

# P branch group (lower wavenumber side)
P_centres = [3.5, 4.0, 4.5]
P_amps    = [0.9, 1.05, 0.80]
for c, a in zip(P_centres, P_amps):
    y += gaussian_peak(x, c, a, 0.08)

# Satellite branches (weak, between and outside main groups)
sat_centres = [2.2, 2.7, 6.0, 6.5, 10.2, 10.7]
sat_amps    = [0.18, 0.22, 0.15, 0.20, 0.14, 0.16]
y_sat = np.zeros_like(x)
for c, a in zip(sat_centres, sat_amps):
    y_sat += gaussian_peak(x, c, a, 0.07)
y += y_sat

ax3.plot(x, y, color='#222222', linewidth=1.5)
ax3.fill_between(x, y, alpha=0.08, color='#1a6faf')

# Colour the main branch groups
for c, a in zip(R_centres, R_amps):
    yg = gaussian_peak(x, c, a, 0.08)
    ax3.fill_between(x, yg, alpha=0.25, color=col_R)
for c, a in zip(P_centres, P_amps):
    yg = gaussian_peak(x, c, a, 0.08)
    ax3.fill_between(x, yg, alpha=0.25, color=col_P)

# Labels
ax3.text(8.5, 1.28, "R$_1$, R$_2$, R$_3$", ha='center', fontsize=9, color=col_R)
ax3.text(4.0, 1.18, "P$_1$, P$_2$, P$_3$", ha='center', fontsize=9, color=col_P)
ax3.text(2.45, 0.32, "Sat.", ha='center', fontsize=8, color=col_sat)
ax3.text(6.25, 0.32, "Sat.", ha='center', fontsize=8, color=col_sat)
ax3.text(10.45, 0.26, "Sat.", ha='center', fontsize=8, color=col_sat)

ax3.set_xlabel(r"Relative wavenumber [cm$^{-1}$]", fontsize=11)
ax3.set_ylabel("Intensity (arb.)", fontsize=11)
ax3.set_ylim(0, 1.45)
ax3.set_xlim(0, 12)
ax3.tick_params(which='major', direction='in', length=5, top=True, right=True, labelsize=9)
ax3.tick_params(which='minor', direction='in', length=3, top=True, right=True)
ax3.xaxis.set_minor_locator(AutoMinorLocator())
ax3.yaxis.set_minor_locator(AutoMinorLocator())
ax3.set_yticks([])
ax3.set_title("Schematic single-band spectrum (fixed v', v'')", fontsize=10)
plt.savefig("pre_1.svg", format="svg", bbox_inches="tight")
print("Done.")
```

```python
# %matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator
import matplotlib.gridspec as gridspec
from matplotlib.lines import Line2D

plt.rcParams.update({"font.family": "serif", "font.size": 12})

col_R   = '#1a6faf'   # blue — R branch
col_P   = '#b22222'   # red  — P branch
col_sat = '#7b3f9e'   # purple — satellite

fig = plt.figure(figsize=(7.5, 11.5))
gs  = gridspec.GridSpec(2, 1, height_ratios=[3.8, 1.0], hspace=0.07)
ax1 = fig.add_subplot(gs[0])   # Grotrian
ax3 = fig.add_subplot(gs[1])   # Spectrum

# ═══════════════════════════════════════════════════════════
# GROTRIAN PANEL
# Convention (matching image):  F3 = J=N+1 (top, highest E)
#                                F2 = J=N   (middle)
#                                F1 = J=N-1 (bottom, lowest E)
# ═══════════════════════════════════════════════════════════
ax1.axis('off')
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 15.2)

# y-centres for each N group
X_base = {1: 1.9,  3: 5.3}   # X state (odd N)
B_base = {2: 9.3,  4: 12.9}  # B state (even N)

# vertical offsets within each N group
dF = {'F3':  0.50,   # top
      'F2':  0.0,    # middle
      'F1': -0.50}   # bottom

x_L, x_R     = 0.08, 0.74
x_Flbl        = 0.76
x_Jlbl        = 0.89

lev_cfg = {
    'F3': dict(color='black',    lw=2.0, ls='-'),
    'F2': dict(color='dimgray',  lw=1.5, ls=(0,(5,3))),
    'F1': dict(color='darkgray', lw=1.2, ls='-'),
}

def draw_N_group(ax, N, base):
    J_map = {'F3': N+1, 'F2': N, 'F1': max(N-1, 0)}
    for fk, cfg in lev_cfg.items():
        y = base + dF[fk]
        ax.hlines(y, x_L, x_R, **cfg, zorder=3)
        ax.text(x_Flbl, y, f'$F_{fk[1]}$', va='center', fontsize=9.5,
                color=cfg['color'])
        ax.text(x_Jlbl, y, f'J={J_map[fk]}', va='center', fontsize=8.5,
                color='gray')
    ax.text(x_L - 0.025, base, f'N={N}',
            va='center', ha='right', fontsize=12, fontweight='bold')

for N, base in X_base.items():
    draw_N_group(ax1, N, base)
for N, base in B_base.items():
    draw_N_group(ax1, N, base)

# State labels
ax1.text(0.01, 11.3, r'B $^3\Sigma_u^+$' + '\n(even N only)',
         va='center', ha='left', fontsize=10.5, color='darkred',
         bbox=dict(fc='#fff5f5', ec='darkred', lw=1.0, boxstyle='round,pad=0.3'))
ax1.text(0.01, 3.6, r'X $^3\Sigma_g^-$' + '\n(odd N only)',
         va='center', ha='left', fontsize=10.5, color='navy',
         bbox=dict(fc='#f0f5ff', ec='navy', lw=1.0, boxstyle='round,pad=0.3'))

# ── Arrow helper ─────────────────────────────────────────────
def arrow(ax, x, y0, y1, col, lw=1.8, ls='-'):
    ax.annotate('', xy=(x, y1), xytext=(x, y0),
                arrowprops=dict(arrowstyle='->', color=col, lw=lw,
                                linestyle=ls, mutation_scale=12))

# ── x positions (left → right) ───────────────────────────────
# P-sat (PQ12,PQ23) | PR13 | P-main | R-main | R-sat (RQ21,RQ32,RP31)
xPs = [0.12, 0.17]              # PQ12, PQ23
xPR =  0.22                     # ᴾR₁₃  ← NEW (was missing)
xPm = [0.30, 0.35, 0.40]       # P1, P2, P3
xRm = [0.48, 0.53, 0.58]       # R1, R2, R3
xRs = [0.65, 0.69, 0.73]       # RQ21, RQ32, RP31

# ── Main R branches: same-F component, N''=1 → N'=2 ─────────
# R1: F1''→F1', R2: F2''→F2', R3: F3''→F3'
# ΔJ = +1 in all cases ✓
for fk, x in zip(['F1', 'F2', 'F3'], xRm):
    arrow(ax1, x, X_base[1]+dF[fk], B_base[2]+dF[fk], col_R, lw=2.2)
ax1.text(np.mean(xRm), (X_base[1]+B_base[2])/2 + 0.55,
         r'R$_1$, R$_2$, R$_3$' + '\n(main)',
         ha='center', va='center', fontsize=8.5, color=col_R,
         bbox=dict(fc='white', ec=col_R, lw=0.6, boxstyle='round,pad=0.25'))

# ── Main P branches: same-F component, N''=3 → N'=2 ─────────
# P1: F1''→F1', P2: F2''→F2', P3: F3''→F3'
# ΔJ = -1 in all cases ✓
for fk, x in zip(['F1', 'F2', 'F3'], xPm):
    arrow(ax1, x, X_base[3]+dF[fk], B_base[2]+dF[fk], col_P, lw=2.2)
ax1.text(np.mean(xPm), (X_base[3]+B_base[2])/2 + 0.55,
         r'P$_1$, P$_2$, P$_3$' + '\n(main)',
         ha='center', va='center', fontsize=8.5, color=col_P,
         bbox=dict(fc='white', ec=col_P, lw=0.6, boxstyle='round,pad=0.25'))

# ── R-type satellite branches (N''=1 → N'=2, ΔN=+1) ─────────
#  RQ₂₁: F2''(N=1,J=1) → F1'(N=2,J=1)   ΔJ=0  ✓
#  RQ₃₂: F3''(N=1,J=2) → F2'(N=2,J=2)   ΔJ=0  ✓
#  ᴿP₃₁: F3''(N=1,J=2) → F1'(N=2,J=1)   ΔJ=-1 ✓
R_sats = [('F2','F1'), ('F3','F2'), ('F3','F1')]
for (flo, fup), x in zip(R_sats, xRs):
    arrow(ax1, x, X_base[1]+dF[flo], B_base[2]+dF[fup], col_sat, lw=1.3, ls='dashed')
ax1.text(np.mean(xRs), (X_base[1]+B_base[2])/2 + 0.9,
         r'RQ$_{21}$, RQ$_{32}$' + '\n' + r'$^R$P$_{31}$ (sat.)',
         ha='center', va='center', fontsize=8, color=col_sat,
         bbox=dict(fc='white', ec=col_sat, lw=0.6, boxstyle='round,pad=0.25'))

# ── P-type satellite branches (N''=3 → N'=2, ΔN=-1) ─────────
#  PQ₁₂: F1''(N=3,J=2) → F2'(N=2,J=2)   ΔJ=0  ✓
#  PQ₂₃: F2''(N=3,J=3) → F3'(N=2,J=3)   ΔJ=0  ✓
P_sats = [('F1','F2'), ('F2','F3')]
for (flo, fup), x in zip(P_sats, xPs):
    arrow(ax1, x, X_base[3]+dF[flo], B_base[2]+dF[fup], col_sat, lw=1.3, ls='dashed')
ax1.text(np.mean(xPs), (X_base[3]+B_base[2])/2 + 0.25,
         r'PQ$_{12}$, PQ$_{23}$' + '\n(sat.)',
         ha='center', va='center', fontsize=8, color=col_sat,
         bbox=dict(fc='white', ec=col_sat, lw=0.6, boxstyle='round,pad=0.25'))

# ── ᴾR₁₃ — THE PREVIOUSLY MISSING SATELLITE ─────────────────
#  ᴾR₁₃: F1''(N=3,J=2) → F3'(N=2,J=3)   ΔN=-1, ΔJ=+1 ✓
#  Connects bottom of N=3 (X) to TOP of N=2 (B): maximally crossing arrow
arrow(ax1, xPR,
      X_base[3]+dF['F1'],   # J=2, bottom of N=3
      B_base[2]+dF['F3'],   # J=3, top of N=2
      col_sat, lw=1.3, ls='dashed')
ax1.text(xPR + 0.005, (X_base[3]+B_base[2])/2 - 0.95,
         r'$^P$R$_{13}$' + '\n(sat.)',
         ha='center', va='center', fontsize=8, color=col_sat,
         bbox=dict(fc='white', ec=col_sat, lw=0.6, boxstyle='round,pad=0.25'))

# ── Legend ───────────────────────────────────────────────────
ax1.legend(handles=[
    Line2D([0],[0], color=col_R,   lw=2,
           label=r'R branch (main, $\Delta N=+1$)'),
    Line2D([0],[0], color=col_P,   lw=2,
           label=r'P branch (main, $\Delta N=-1$)'),
    Line2D([0],[0], color=col_sat, lw=1.5, ls='--',
           label=r'Satellite branches ($\Delta F\neq\Delta N$)'),
], loc='upper left', fontsize=8.5, framealpha=0.93,
   bbox_to_anchor=(0.24, 0.998))

# ═══════════════════════════════════════════════════════════
# SPECTRUM PANEL
# ᴾR₁₃ spectrally: ΔN=-1 but ΔJ=+1 from lowest X sub-level
# to highest B sub-level → higher ν than PQ satellites,
# falls between P-group inner edge and R-group outer edge
# ═══════════════════════════════════════════════════════════
def gauss(x, c, a, w):
    return a * np.exp(-0.5 * ((x - c) / w) ** 2)

xv = np.linspace(0, 12, 3000)
y  = np.zeros_like(xv)

# Main R branch: three peaks
for c, a in zip([8.0, 8.5, 9.0], [1.0, 1.15, 0.85]):
    y += gauss(xv, c, a, 0.08)
# Main P branch: three peaks
for c, a in zip([3.5, 4.0, 4.5], [0.9, 1.05, 0.80]):
    y += gauss(xv, c, a, 0.08)

# Satellite peaks:
#   PQ12, PQ23 → far left (~2.0, 2.5)
#   ᴾR₁₃      → between P and R groups (~5.5)  ← NEW distinct peak
#   RQ21, RP31 → far right (~10.0, 10.6)
sat_peaks = [(2.0, 0.18), (2.5, 0.22),
             (5.5, 0.21),               # ᴾR₁₃
             (10.0, 0.16), (10.6, 0.14)]
for c, a in sat_peaks:
    y += gauss(xv, c, a, 0.07)

ax3.plot(xv, y, color='#222222', lw=1.4)
ax3.fill_between(xv, y, alpha=0.07, color='#555555')

# Colour the main branches
for c, a in zip([8.0, 8.5, 9.0], [1.0, 1.15, 0.85]):
    ax3.fill_between(xv, gauss(xv, c, a, 0.08), alpha=0.25, color=col_R)
for c, a in zip([3.5, 4.0, 4.5], [0.9, 1.05, 0.80]):
    ax3.fill_between(xv, gauss(xv, c, a, 0.08), alpha=0.25, color=col_P)

# Labels
ax3.text(8.5,  1.28, r'R$_1$, R$_2$, R$_3$',  ha='center', fontsize=9,   color=col_R)
ax3.text(4.0,  1.18, r'P$_1$, P$_2$, P$_3$',  ha='center', fontsize=9,   color=col_P)
ax3.text(2.25, 0.36, 'Sat.',                   ha='center', fontsize=8.5, color=col_sat)
ax3.text(5.5,  0.35, r'$^P$R$_{13}$',          ha='center', fontsize=8.5, color=col_sat)
ax3.text(10.3, 0.27, 'Sat.',                   ha='center', fontsize=8.5, color=col_sat)

ax3.set_xlabel(r'Relative wavenumber [cm$^{-1}$]', fontsize=11)
ax3.set_ylabel('Intensity (arb.)', fontsize=11)
ax3.set_xlim(0, 12)
ax3.set_ylim(0, 1.50)
ax3.tick_params(which='major', direction='in', length=5, top=True, right=True)
ax3.tick_params(which='minor', direction='in', length=3, top=True, right=True)
ax3.xaxis.set_minor_locator(AutoMinorLocator())
ax3.yaxis.set_minor_locator(AutoMinorLocator())
ax3.set_yticks([])
ax3.set_title("Schematic single-band spectrum (fixed v', v'')", fontsize=10)
plt.show()

# plt.savefig('/mnt/user-data/outputs/sr_branch_structure.svg',
#             format='svg', bbox_inches='tight', dpi=200)
# plt.savefig('/mnt/user-data/outputs/sr_branch_structure.png',
#             format='png', bbox_inches='tight', dpi=200)
print("Done.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

# --- 1. Aesthetic Setup ---
plt.rcParams.update({
    "font.family": "serif",
    "axes.linewidth": 2.0,
    "figure.dpi": 150
})

fig, ax = plt.subplots(figsize=(4,3))
ax.axis('off')  # Turn off standard plotting axes completely

# --- 2. Defining qualitative oxygen potentials ---
R = np.linspace(0.8, 7.5, 500)

# (a) Ground state O2 well (asymptotes to O(3P)+O(3P))
V_ground = 3.5 * (1 - np.exp(-1.5 * (R - 2.0)))**2 + 0.5
V_ground[R > 4] = 2.0 + (V_ground[R > 4] - 2.0) * np.exp(-(R[R > 4] - 4))

# (b) Excited Schumann-Runge-like bound state
V_excited = 2.5 * (1 - np.exp(-1.2 * (R - 2.2)))**2 + 2.5
V_excited[R > 4.5] = 4.5 + (V_excited[R > 4.5] - 4.5) * np.exp(-(R[R > 4.5] - 4.5))

# (c) Repulsive Pi-state curve crossing the excited manifold
V_repulsive = 10.0 * np.exp(-2.0 * (R - 1.5)) + 2.0

# Plot curves with distinct colors
ground_color = '#3b2f2f'
excited_color = '#8b1a1a'
pi_color = '#1f6f8b'

ax.plot(R, V_ground, color=ground_color, lw=3, zorder=3)
ax.plot(R, V_excited, color=excited_color, lw=3, zorder=3)
ax.plot(R+0.3, V_repulsive, color=pi_color, lw=3, zorder=3)

# --- 3. Vibrational levels in excited bound state ---
levels = [2.9, 3.2, 3.5]
for E in levels:
    mask = V_excited < E
    R_level = R[mask]
    ax.plot([R_level[0], R_level[-1]], [E, E], color=excited_color, lw=2, zorder=2)

# --- 4. Arrows (excitation and predissociation pathway) ---
arrow_color = '#c0392b'

# Vertical photoexcitation arrow
ax.annotate('', xy=(2.0, 3.5), xytext=(2.0, 0.5),
            arrowprops=dict(facecolor=arrow_color, edgecolor=arrow_color,
                            width=2, headwidth=8, headlength=10), zorder=4)

# Curved predissociation arrow toward repulsive channel
curved_arrow = FancyArrowPatch((2.8, 3.6), (4.5, 2),
                               connectionstyle="arc3,rad=-0.3",
                               color=arrow_color, lw=2.5,
                               arrowstyle='-|>,head_width=6,head_length=8', zorder=4)
ax.add_patch(curved_arrow)

# Dissociation-energy guide
ax.plot([2.0, 7.5], [0.5, 0.5], 'k--', lw=1, alpha=0.5)
ax.annotate('', xy=(7.0, 2.0), xytext=(7.0, 0.5),
            arrowprops=dict(arrowstyle='<->', color='black', lw=2))
ax.text(6.8, 1.2, '$D_e$', fontsize=14, va='center', ha='right')

# --- 5. Oxygen-relevant labels ---
ax.text(7.0, 2.08, r'O($^3P$) + O($^3P$)', fontsize=12, color=ground_color)
ax.text(6.95, 4.58, r'O($^1D$) + O($^3P$)', fontsize=12, color=excited_color)
ax.text(2.05, -0.02, r'O$_2$ (X $^3\Sigma_g^-$)', fontsize=12, ha='center')

ax.text(4.2, 4.95, r'B $^3\Sigma_u^+$ (SR)', fontsize=10, color=excited_color)
ax.text(2.15, 3.95, r'Repulsive $\Pi$ state', fontsize=10, color=pi_color, rotation=-80)

ax.set_ylim(-0.5, 8.0)
ax.set_xlim(0.5, 8.5)
plt.savefig("pre_2.svg", format="svg", bbox_inches="tight")

plt.tight_layout()
plt.show()
```

```python
"""
Figure 1.2 (thesis Fig 1.1): Non-equilibrium population comparison
Fixed version:
  - alpha_eff/Error boxes no longer clipped in panels B, C
  - Panel B annotation arrow points at v=0 (the actual peak), describes spread
  - Panel D arrows no longer converge to same point
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
import matplotlib.patches as mpatches
import matplotlib.cm as cm

# ── global style ──────────────────────────────────────────────────────────────
plt.rcParams.update({
    "font.family":        "serif",
    "font.size":          13,
    "axes.labelsize":     13,
    "axes.titlesize":     13,
    "xtick.labelsize":    11.5,
    "ytick.labelsize":    11.5,
    "axes.linewidth":     1.4,
    "xtick.direction":    "in",
    "ytick.direction":    "in",
    "xtick.major.size":   5,
    "ytick.major.size":   5,
    "xtick.minor.size":   3,
    "ytick.minor.size":   3,
    "xtick.major.width":  1.2,
    "ytick.major.width":  1.2,
    "legend.fontsize":    11,
    "legend.framealpha":  0.92,
    "legend.edgecolor":   "0.7",
})

# ── physics ───────────────────────────────────────────────────────────────────
k_B_cm = 0.6950356

def E_vib(v):
    return 1580.*(v + 0.5) - 12.0*(v + 0.5)**2

def boltzmann(T, n=18):
    E = np.array([E_vib(v) - E_vib(0) for v in range(n)])
    w = np.exp(-E / (k_B_cm * T))
    return w / w.sum()

N_V = 18
v_arr = np.arange(N_V)

# state-specific alpha relative to v=0 at 633 nm
alpha_rel = np.array([1.000, 0.619, 0.373, 0.306, 0.228, 0.169,
                       0.151, 0.115, 0.104, 0.091, 0.076, 0.073,
                       0.061, 0.059, 0.052, 0.048, 0.044, 0.042])

pop_A = boltzmann(300)                            # (A) equilibrium 300 K
pop_B = boltzmann(3000)                           # (B) T_vib = 3000 K
pop_C = np.exp(-0.5 * ((v_arr - 5) / 1.8)**2)    # (C) selective v=5
pop_C /= pop_C.sum()

def alpha_eff(pop):
    return sum(pop[v] * alpha_rel[v] for v in range(N_V))

aA = alpha_eff(pop_A)
aB = alpha_eff(pop_B)
aC = alpha_eff(pop_C)

err_B = (aB - aA) / aA * 100.   # −25 %
err_C = (aC - aA) / aA * 100.   # −80 %

# ── colour palette ────────────────────────────────────────────────────────────
CA = "#1a3a6b"    # blue  – equilibrium
CB = "#8b1a1a"    # red   – two-temperature
CC = "#226633"    # green – selective
CGRAY = "#888888"

# ── figure layout ─────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(15, 10))
fig.patch.set_facecolor("white")

gs = gridspec.GridSpec(
    2, 3,
    figure=fig,
    hspace=0.54,
    wspace=0.32,
    top=0.95, bottom=0.09,
    left=0.07, right=0.97,
)

ax_A  = fig.add_subplot(gs[0, 0])
ax_B  = fig.add_subplot(gs[0, 1])
ax_C  = fig.add_subplot(gs[0, 2])
ax_D  = fig.add_subplot(gs[1, :])   # full-width bottom row

POP_PANELS = [
    (ax_A, pop_A, CA,
     r"Equilibrium — $T = 300\,\mathrm{K}$",
     "Boltzmann distribution\n(laboratory / quiescent gas)",
     aA, 0.0,
     r"$v''=0$ carries 99.9 %""\nof population", 0),

    (ax_B, pop_B, CB,
     r"Two-Temperature — $T_{vib}=3000\,\mathrm{K},\;T_{rot}=300\,\mathrm{K}$",
     "Post-shock / laser-heated gas\n(vibrational and rotational modes decouple)",
     aB, err_B,
     r"Higher $v''$ populated;" "\nBoltzmann spread,"
     "\n"r"$T_{vib}\neq T_{rot}$", 3),

    (ax_C, pop_C, CC,
     r"Selective Excitation — peak at $v''=5$",
     "Laser pumping / expansion cooling\n(non-Boltzmann, peaked distribution)",
     aC, err_C,
     r"Non-Boltzmann:" "\n" r"$v''=5$ dominates", 5),
]

# ── population bar panels ─────────────────────────────────────────────────────
bar_cmap = cm.get_cmap("viridis_r")

for ax, pop, col, title, subtitle, a_val, err, annot_txt, annot_v in POP_PANELS:

    # coloured bars
    for v in range(N_V):
        bar_col = bar_cmap(v / (N_V - 1))
        ax.bar(v, pop[v] * 100.,
               width=0.72, color=bar_col, alpha=0.88,
               edgecolor="white", linewidth=0.6, zorder=3)

    # axes cosmetics — set limits BEFORE placing annotations
    ax.set_xlabel(r"Vibrational level $v''$", fontsize=13)
    ax.set_xlim(-0.65, N_V - 0.35)
    y_top = min(pop.max() * 100. * 1.30, 125.)
    ax.set_ylim(0, y_top)
    ax.xaxis.set_major_locator(MultipleLocator(3))
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(which="major", direction="in", length=5,
                   top=True, right=True)
    ax.tick_params(which="minor", direction="in", length=3,
                   top=True, right=True)

    # alpha annotation box — placed INSIDE the axes area with clip_on=False
    err_str   = f"{err:+.0f} %" if err != 0. else "reference"
    err_color = "#b22222" if err < -5 else ("#226633" if err > 5 else "#555")
    ax.text(0.97, 0.93,
            r"$\alpha_\mathrm{eff} = $" + f"{a_val:.3f}" + r"$\,\alpha_0$"
            + "\n" + f"Error: {err_str}",
            ha="right", va="top", transform=ax.transAxes,
            fontsize=11.5, color=err_color, fontweight="bold",
            clip_on=False,
            bbox=dict(fc="white", ec=err_color, lw=1.0,
                      boxstyle="round,pad=0.30", alpha=0.95))

    # annotation arrow pointing to relevant level
    if pop[annot_v] > 0.005:
        # For panel B: arrow points at v=3 region to highlight spread
        # Place text to the right, in clear space
        if annot_v == 3:
            # Panel B: point at the v=3 bar, text to the right in empty space
            txt_x = annot_v + 4.5
            txt_y = pop[annot_v] * 100. + 12
        elif annot_v == 0:
            # Panel A: point at v=0, text offset to right
            txt_x = annot_v + 2.8
            txt_y = pop[annot_v] * 100. * 0.55 + 6
        else:
            # Panel C: point at v=5
            txt_x = annot_v + 3.5
            txt_y = pop[annot_v] * 100. * 0.55 + 6

        ax.annotate(annot_txt,
                    xy=(annot_v, pop[annot_v] * 100. + 0.8),
                    xytext=(txt_x, txt_y),
                    fontsize=10.5, color=col,
                    arrowprops=dict(arrowstyle="->", color=col,
                                    lw=1.3, mutation_scale=12),
                    bbox=dict(fc="white", ec=col, lw=0.8,
                              boxstyle="round,pad=0.22", alpha=0.93))

    # title and subtitle
    ax.set_title(title, fontsize=12.5, fontweight="bold", pad=6, color=col)
    ax.text(0.5, -0.23, subtitle,
            ha="center", va="top", transform=ax.transAxes,
            fontsize=10.5, color="#555", style="italic", linespacing=1.45)

ax_A.set_ylabel("Population (%)", fontsize=13)

# ── diagnostic implication row (D) ───────────────────────────────────────────
ax = ax_D
ax.set_xlim(0, 14)
ax.set_ylim(0, 4.8)
ax.axis("off")
ax.set_facecolor("white")

ax.text(7.0, 4.62,
        "Why Existing Data Fails: The State-Specificity Requirement",
        ha="center", va="top", fontsize=13.5,
        fontweight="bold", color="#1a1a1a")

# ── three scenario columns ────────────────────────────────────────────────────
COL_X  = [2.3,  7.0,  11.7]
SC_COL = [CA,   CB,   CC]
SC_LBL = ["Eq. 300 K",
          r"$T_{vib}=3000\,\mathrm{K}$",
          r"Selective $v''=5$"]
ALPHA_LBL = [r"$\alpha_{300\mathrm{K}}$  (correct)",
             r"$\alpha_{300\mathrm{K}}$  (wrong!)",
             r"$\alpha_{300\mathrm{K}}$  (wrong!)"]
OUTCOME_TXT = ["$N$ retrieved\ncorrectly",
               "$N$ over-estimated\nby 34 %",
               "$N$ over-estimated\nby 79 %"]
OUTCOME_COL = ["#226633", "#b22222", "#b22222"]

BOX_W = 3.15
ROW_Y_TOP = 3.55   # population boxes
ROW_Y_BOT = 1.55   # outcome boxes

def rbox(ax, cx, cy, w, h, col, lw=1.6, fa=0.15):
    p = FancyBboxPatch((cx - w/2, cy - h/2), w, h,
                        boxstyle="round,pad=0.18",
                        facecolor=col, alpha=fa,
                        edgecolor=col, linewidth=lw, zorder=3)
    ax.add_patch(p)

def dn_arrow(ax, cx, y1, y2, col, lw=1.6):
    ax.annotate("", xy=(cx, y2), xytext=(cx, y1),
                arrowprops=dict(arrowstyle="->", color=col,
                                lw=lw, mutation_scale=13), zorder=5)

for cx, col, slbl, albl, otxt, ocol in zip(
        COL_X, SC_COL, SC_LBL, ALPHA_LBL, OUTCOME_TXT, OUTCOME_COL):

    # population box
    rbox(ax, cx, ROW_Y_TOP, BOX_W, 0.88, col, fa=0.14)
    ax.text(cx, ROW_Y_TOP, slbl,
            ha="center", va="center", fontsize=12,
            fontweight="bold", color=col, zorder=6)

    # arrow + alpha label
    dn_arrow(ax, cx, ROW_Y_TOP - 0.44, ROW_Y_BOT + 0.50, "#555")
    ax.text(cx + 0.68, (ROW_Y_TOP + ROW_Y_BOT) / 2 + 0.02,
            albl, ha="left", va="center",
            fontsize=11, color="#555", style="italic")

    # outcome box
    rbox(ax, cx, ROW_Y_BOT, BOX_W, 0.88, ocol, fa=0.14)
    ax.text(cx, ROW_Y_BOT, otxt,
            ha="center", va="center", fontsize=12,
            fontweight="bold", color=ocol, zorder=6,
            linespacing=1.4)

    # arrow down from outcome
    dn_arrow(ax, cx, ROW_Y_BOT - 0.44, 0.82, ocol, lw=1.4)

# solution box at the bottom
SOL_BOX = FancyBboxPatch((3.0, 0.10), 8.0, 0.68,
                           boxstyle="round,pad=0.20",
                           facecolor="#b22222", alpha=0.12,
                           edgecolor="#b22222", linewidth=2.2, zorder=3)
ax.add_patch(SOL_BOX)
ax.text(7.0, 0.45,
        r"Solution: $\alpha(v'', J'', \lambda_s)$ — "
        "state-specific, population-independent spectroscopic database",
        ha="center", va="center", fontsize=11.5,
        fontweight="bold", color="#b22222", zorder=6)

# arrows from the two wrong outcomes toward the solution box
# Fixed: each arrow lands at a distinct point on the solution box
ax.annotate("", xy=(7.0, 0.78), xytext=(7.0, 1.11),
            arrowprops=dict(arrowstyle="->", color="#b22222",
                            lw=1.2, mutation_scale=11), zorder=5)

ax.annotate("", xy=(9.5, 0.78), xytext=(11.7, 1.11),
            arrowprops=dict(arrowstyle="->", color="#b22222",
                            lw=1.2, mutation_scale=11,
                            connectionstyle="arc3,rad=-0.15"), zorder=5)

# row labels on the far left
ax.text(0.18, ROW_Y_TOP,
        "Population\ndistribution", ha="left", va="center",
        fontsize=11, color=CGRAY, style="italic", linespacing=1.4)
ax.text(0.18, ROW_Y_BOT,
        "Diagnostic\noutcome", ha="left", va="center",
        fontsize=11, color=CGRAY, style="italic", linespacing=1.4)

plt.savefig("/storage/home/hcoda1/9/pagrawala6/r-js130-0/thesisplots/newfolder2/Ch1_1.svg", format="svg",
            dpi=400, bbox_inches="tight", facecolor="white")
print("Saved.")
```
