```python
import numpy as np
import matplotlib.pyplot as plt
J_prime = np.linspace(2, 50, 49, dtype=int)  # Natural numbers from 2 to 70
nu_array = np.linspace(50022, 52008, 49)
# Declare Einstein A variables (placeholder functions - to be modified later)
c = 1


# putting in nu data for the v' = 10 to v" = 2 band
def strength_P(J):
    nu = -((52022.6106 - 50290.19) / 47)*(J-2) + 52008.563
    FCF_dependence = 1 - 0.0127*(J-1)
    return (J - (1/J))  * c * nu**3  * FCF_dependence

def strength_R(J):
    nu = -((52008.563 - 50022.21) / 47)*(J-2) + 52008.563
    FCF_dependence = 1 - 0.0127*(J-1)
    return (J+1 - (1/(J+1))) * c * nu**3  * FCF_dependence




# Calculate Einstein A coefficients
A_Pbranch = strength_P(J_prime) / (2*J_prime + 1)
A_Rbranch = strength_R(J_prime) / (2*J_prime + 1)




plt.figure(figsize=(7, 5))
plt.ylim(2e13,8e13)
plt.plot(J_prime, A_Pbranch, 'bo-', linewidth=2, markersize=4, label='P-branch', alpha=0.8)
plt.plot(J_prime, A_Rbranch, 'ro-', linewidth=2, markersize=4, label='R-branch', alpha=0.8)
plt.xlabel(r"$J'$", fontsize=14, fontweight='bold')
plt.ylabel(r"Einstein A", fontsize=14, fontweight='bold')
plt.title(r"$A_{J''}^{J'}$ vs Upper Quantum No. $J'$ (J, $\nu$, estimated FCF dependence)", fontsize=12, fontweight='bold')
plt.grid(False)
plt.minorticks_off()
plt.tick_params(axis='both', which='major', labelsize=12, width=1.5, length=6)
plt.legend(fontsize=12, frameon=True, fancybox=True, shadow=True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# ── Global style ─────────────────────────────────────────────────────────────
plt.rcParams.update({
    'font.family':       'serif',
    'font.serif':        ['DejaVu Serif', 'Times New Roman', 'Times', 'serif'],
    'font.size':         12,
    'axes.linewidth':    1.2,
    'xtick.direction':   'in',
    'ytick.direction':   'in',
    'xtick.top':         True,
    'ytick.right':       True,
    'xtick.major.size':  6,
    'xtick.minor.size':  3,
    'ytick.major.size':  6,
    'ytick.minor.size':  3,
    'xtick.major.width': 1.1,
    'ytick.major.width': 1.1,
    'xtick.minor.width': 0.8,
    'ytick.minor.width': 0.8,
    'legend.frameon':    True,
    'legend.framealpha': 0.9,
    'legend.edgecolor':  '0.3',
    'legend.fontsize':   11,
    'figure.dpi':        150,
})

# ── Data ─────────────────────────────────────────────────────────────────────
J_prime = np.linspace(2, 50, 49, dtype=int)
c = 1

# ── Model 1: HITRAN (ν + Hönl–London only) ──────────────────────────────────
def strength_P_hitran(J):
    nu = -((52022.6106 - 50290.19) / 47) * (J - 2) + 52008.563
    return (J - 1/J) * c * nu**3

def strength_R_hitran(J):
    nu = -((52008.563 - 50022.21) / 47) * (J - 2) + 52008.563
    return (J + 1 - 1/(J + 1)) * c * nu**3

A_P_hitran = strength_P_hitran(J_prime) / (2*J_prime + 1)
A_R_hitran = strength_R_hitran(J_prime) / (2*J_prime + 1)

# ── Model 2: Rotation–vibration coupling (+ J-dependent FCF) ─────────────────
def strength_P_rvc(J):
    nu = -((52022.6106 - 50290.19) / 47) * (J - 2) + 52008.563
    FCF = 1 - 0.0127 * (J - 1)
    return (J - 1/J) * c * nu**3 * FCF

def strength_R_rvc(J):
    nu = -((52008.563 - 50022.21) / 47) * (J - 2) + 52008.563
    FCF = 1 - 0.0127 * (J - 1)
    return (J + 1 - 1/(J + 1)) * c * nu**3 * FCF

A_P_rvc = strength_P_rvc(J_prime) / (2*J_prime + 1)
A_R_rvc = strength_R_rvc(J_prime) / (2*J_prime + 1)

# ── Figure ────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7.5, 5.2))

# Colour  → branch.   Marker/linestyle → model.
R_col = '#C0392B'   # red  — R-branch
P_col = '#1A5276'   # blue — P-branch

# HITRAN: solid line, filled circle
ax.plot(J_prime, A_R_hitran, '-o',
        color=R_col, linewidth=2.0, markersize=6,
        markerfacecolor=R_col, markeredgewidth=0.5, markeredgecolor='white',
        label=r'$R$-branch  (HITRAN model)', zorder=4)
ax.plot(J_prime, A_P_hitran, '-o',
        color=P_col, linewidth=2.0, markersize=6,
        markerfacecolor=P_col, markeredgewidth=0.5, markeredgecolor='white',
        label=r'$P$-branch  (HITRAN model)', zorder=4)

# Rot.–vib. coupling: dashed line, open square
ax.plot(J_prime, A_R_rvc,
        color=R_col, linewidth=1.8, markersize=6,
        marker='s', markerfacecolor='none', markeredgewidth=1.4, markeredgecolor=R_col,
        linestyle=(0, (5, 2)),
        label=r'$R$-branch  (rot.–vib. coupling)', zorder=3)
ax.plot(J_prime, A_P_rvc,
        color=P_col, linewidth=1.8, markersize=6,
        marker='s', markerfacecolor='none', markeredgewidth=1.4, markeredgecolor=P_col,
        linestyle=(0, (5, 2)),
        label=r'$P$-branch  (rot.–vib. coupling)', zorder=3)

# ── Axes ──────────────────────────────────────────────────────────────────────
ax.set_xlim(0, 52)
ax.set_ylim(2e13, 8e13)

ax.set_xlabel(r"$J'$", fontsize=13)
ax.set_ylabel(r"Einstein $A$ coefficient  [s$^{-1}$]", fontsize=13)

ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(2))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1e13))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(5))

# Scalar tick labels + explicit multiplier annotation
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/1e13:.0f}'))
ax.text(0.0, 1.01, r'$\times 10^{13}$', transform=ax.transAxes,
        fontsize=11, ha='left', va='bottom')

# ── Legend ────────────────────────────────────────────────────────────────────
leg = ax.legend(loc='lower left', handlelength=2.4, handleheight=1.0,
                labelspacing=0.45, borderpad=0.7)
leg.get_frame().set_linewidth(0.8)

fig.tight_layout()
# fig.savefig('/mnt/user-data/outputs/einstein_A_branches.pdf', bbox_inches='tight')
# fig.savefig('/mnt/user-data/outputs/einstein_A_branches.png', dpi=200, bbox_inches='tight')
plt.savefig('/storage/home/hcoda1/9/pagrawala6/r-js130-0/thesisplots/newfolder2/einstein_A_branches.svg', format='svg', bbox_inches='tight')
print("Saved.")
```
